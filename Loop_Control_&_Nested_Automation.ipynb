```python
#Task 1 – Nested Loop for Multi-Class Grade Analysis
#Step 1 – School Data
school_data = [
    ["Alice", 85, 92, 78],      # [name, math, science, english]
    ["Bob", 72, 68, 81],
    ["Charlie", 95, 88, 91],
    ["Diana", 63, 59, 67],
    ["Eve", 88, 91, 85]
]
```

```python
#Step 2 – Use Nested Loops
for student in school_data:      # Outer loop for each student

    name = student[0]
    total = 0

    # Inner loop for subject scores
    for score in student[1:]:
        total += score

    average = total / 3

    if average < 70:
        continue

    print(f"{name}: Average = {average} - PASS")
```

# Outer loop iterates through each student in the nested list.
# Inner loop iterates through the subject scores.
# The continue statement skips students whose average is below 70.

```python
#Task 2 – Tournament Bracket with Break Control
#Step 1 – Tournament Data
tournament = [
    ["Player_A1", "Player_A2", "Player_A3"],  # Advanced level
    ["Player_B1", "Player_B2", "Player_B3"],  # Beginner level
    ["Player_I1", "Player_I2", "Player_I3"]   # Intermediate level
]
```

```python
#Step 2 – Nested Loops with Break
stop_tournament = False

for group in tournament:   # Outer loop for skill groups

    for i in range(len(group) - 1):   # Inner loop for matches

        player1 = group[i]
        player2 = group[i + 1]

        # Check for special player
        if "I2" in player1 or "I2" in player2:
            print("Tournament paused - Special player found")
            stop_tournament = True
            break

        print(f"Match: {player1} vs {player2}")

    if stop_tournament:
        break
```

# Outer loop iterates through each skill group.
# Inner loop creates matches between players in the same group.
# The break statement stops the loop when a player with "I2" is found.
# A flag variable is used to break out of both loops.


```python
#Task 3 – While Loop Number Pattern
#Step 1 – While Loop Code
number = 1

while number <= 20:

    if number % 3 == 0 and number % 5 == 0:
        number += 1
        continue

    elif number % 3 == 0:
        print(f"{number}: Fizz")

    elif number % 5 == 0:
        print(f"{number}: Buzz")

    else:
        print(number)

    number += 1
```

# The while loop runs from 1 to 20.
# continue is used to skip numbers divisible by both 3 and 5.
# If a number is divisible by 3 it prints "Fizz".
# If a number is divisible by 5 it prints "Buzz".
